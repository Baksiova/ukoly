```python
import numpy as np

def find_print_outliers(df_orders, books_count):
    Q1 = df_orders[books_count].quantile(0.25)
    Q3 = df_orders[books_count].quantile(0.75)
    IQR = Q3 - Q1
    
    # Definovanie outliers
    outliers = df_orders[(df_orders[books_count] < (Q1 - 1.5 * IQR)) | (df_orders[books_count] > (Q3 + 1.5 * IQR))]
    
    if not outliers.empty:
        print(f"Outliers in {books_count}:")
        print(outliers)
    else:
        print(f"No outliers in {books_count}")

find_print_outliers(df_orders, 'total_price')

find_print_outliers(df_orders, 'books_count')


if not np.issubdtype(df_orders['date'].dtype, np.datetime64):
    df_orders['date'] = pd.to_datetime(df_orders['date'], unit='ms')
find_print_outliers(df_orders, 'date')
```

```python
#uprava books_count riadkov u zrusenych objednavok s nenulovou hodnotou


#Počet riadkov so statusom 'status_canceled' rovným 1
cancelled_count = (df_merged['status_canceled'] == 1).sum()

#Počet riadkov, kde 'books_count' má hodnotu NaN
nan_books_count = df_merged['books_count'].isna().sum()

#rozdiel
mismatch = cancelled_count - nan_books_count

print(f"Počet riadkov so status_canceled = 1: {cancelled_count}")
print(f"Počet riadkov s NaN v books_count: {nan_books_count}")
print(f"Rozdiel: {mismatch}")

#Riadky so statusom zrušené
status_canceled_rows = df_merged[df_merged['status_canceled'] == 1]

#Riadky, kde 'books_count' je NaN
nan_books_count_rows = df_merged[df_merged['books_count'].isna()]

#riadky,ktoré  sú v 'status_canceled_rows', ale nie v 'nan_books_count_rows'
mismatch_rows = status_canceled_rows[~status_canceled_rows.index.isin(nan_books_count_rows.index)]

# books_count' na NaN pre riadky, kde je rozdiel
df_merged.loc[mismatch_rows.index, 'books_count'] = np.nan
```
